Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)

X_train_scaled=scaler.transform(X_train)

X_test_scaled=scaler.transform(X_test)


print(X_train_scaled)
X_train_scaled.shape

#print(X_test_scaled)

#X_test_scaled.shape
```

Model

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline

```

```python
model=Sequential()
model.add(Dense(128,input_dim=n+2,activation='relu'))
model.add(Dense(128,activation='relu'))



#output layer

model.add(Dense(2*n,activation='linear'))

model.compile(loss='mean_squared_error',optimizer='adam',metrics=['mae'])

model.summary()

```

Fit the model

```python
history=model.fit(X_train_scaled,y_train,validation_split=0.2,epochs=300)
```

```python
loss=history.history['loss']
val_loss=history.history['val_loss']
epochs=range(1,len(loss)+1)
plt.plot(epochs,loss,'y',label='training loss')


plt.plot(epochs,val_loss,'r',label='validation loss')


plt.title('Training and validation loss')
plt.xlabel('epochs') # epochs is the numb of times the model is trained
plt.ylabel('loss') # loss is abst error
plt.legend()

plt.show()
```

```python
print(history.history.keys())
```

```python
acc=history.history['mae']


val_acc=history.history['val_mae']
epochs=range(1,len(loss)+1)
plt.plot(epochs,acc,'y',label='training mae')


plt.plot(epochs,val_acc,'r',label='validation mae')


plt.title('Training and validation mae')
plt.xlabel('epochs') # epochs is the numb of times the model is trained
plt.ylabel('accuracy') # loss is abst error
plt.legend()

plt.show()
```

Other models

```python
#Linear Regressor

from sklearn import linear_model
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error


lr_model=linear_model.LinearRegression()
lr_model.fit(X_train_scaled,y_train)
y_predict_lr= lr_model.predict(X_test_scaled)
mse_lr=mean_squared_error(y_test,y_predict_lr)
mae_lr=mean_absolute_error(y_test,y_predict_lr)

print(mse_lr)
print(mae_lr)
```

```python
# Decision tree

tree=DecisionTreeRegressor()
tree.fit(X_train_scaled,y_train)

y_predict_tree= tree.predict(X_test_scaled)
mse_tree=mean_squared_error(y_test,y_predict_tree)
mae_tree=mean_absolute_error(y_test,y_predict_tree)

print(mse_tree)
print(mae_tree)
```

```python
# Random forest

from sklearn.ensemble import RandomForestRegressor
model_tree=RandomForestRegressor(n_estimators=10,random_state=30)

model_tree.fit(X_train_scaled,y_train)


y_predict_randomforest= model_tree.predict(X_test_scaled)
mse_rm=mean_squared_error(y_test,y_predict_randomforest)
mae_rm=mean_absolute_error(y_test,y_predict_randomforest)

print(mse_rm)
print(mae_rm)


#Feature ranking

feature_list=list(X_train.columns)
feature_imp=pd.Series(model_tree.feature_importances_,index=feature_list).sort_values(ascending=False)
print(feature_imp)
```
